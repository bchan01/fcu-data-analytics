```python
# Unpivot DataFrame to get Account/Value Pairs
# List of base columns you want to preserve
base_columns = ["CU_NUMBER", "CYCLE_DATE", "JOIN_NUMBER", "UPDATE_DATE"]
# List of columns to melt (all columns that are not base columns)
account_columns = [col for col in fs220.columns if col not in base_columns]
# Melt the DataFrame
df = fs220.melt(id_vars=base_columns,  # Correctly using id_vars
                    value_vars=account_columns,  # Columns to melt
                    var_name='account',  # The name for the variable column
                    value_name='value'  # The name for the value column
                   )
df_1 = df[df['CU_NUMBER'] == 1]
df_1
```
